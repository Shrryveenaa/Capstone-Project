```python
page=requests.get("https://www.skechers.com/")
page
```

```python
soup=BeautifulSoup(page.text,'html.parser')
```

```python
shoe_elements = soup.find_all("div",class_="product b-product")
shoe_elements
```

```python
names=[]
category=[]
prices=[]
colors=[]

for shoe in shoe_elements:
    Shoenames = shoe.find('div', class_="pdp-link c-product-tile__title__wrap").text.strip()
    Category = shoe.find('div', class_="c-product-tile__gender").text.strip()
    Price = shoe.find('span', class_="sales").text.strip()
    Color = shoe.find('div', class_="color-swatches c-product-tile__color-swatches").text.strip().replace('\nAlso in Wide','').replace('\n','')
    
    names.append(Shoenames)
    category.append(Category)
    prices.append(Price)
    colors.append(Color)
```

```python
shoes=({'ShoesName':names,'Category':category,'Price':prices,'Colors':colors})
```

```python
data= pd.DataFrame(shoes)
data
```

```python
csv_file='sTable1.csv'
data.to_csv(csv_file, index=False)
```
