```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page=requests.get("https://www.columbia.com/c/gift-guide/?prefv1=Shoes&prefn1=productClass&icpa=flp&icid=subhero&icsa=HOL&prid=giftguide&icst=nonpr&crid=giftguide-shopfootwear&icca=img")
page
```

```python
soup=BeautifulSoup(page.text,'html.parser')

shoe_elements= soup.find_all("div",class_="product")
shoe_elements
```

```python
names=[]
category=[]
prices=[]
colors=[]


for shoe in shoe_elements:
    Shoenames = shoe.find('h2',class_="product-tile__title pdp-link").text.strip()
    Category = shoe.find('h2',class_="product-tile__title pdp-link")
    Price = shoe.find('span',class_="sales").text.strip().replace('Regular price:','').replace('\n','').replace('Sale price:','')
    Color = shoe.find('div',class_="color-swatches").text.strip().replace('\n','')
    
    names.append(Shoenames)
    category.append(Category.text.strip() if Category else "")
    prices.append(Price)
    colors.append(Color)
    

shoes={'ShoesName':names,'Category':category,'Price':prices,'Colors':colors}
data= pd.DataFrame(shoes)

data['Category'] = data['Category'].str.extract('(Men|Women)', expand=False) 
data['Colors'] = data['Colors'].apply(lambda x: len(x.split(',')))

print(data)
```

```python
csv_file='cTable.csv'
data.to_csv(csv_file, index=False)
```
